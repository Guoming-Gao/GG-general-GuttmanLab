```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True, style="white")
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
folder = "/Users/GGM/Library/CloudStorage/Dropbox/Caltech_PostDoc_GuttmanLab/Conferences/RNA2025"
fname = "quick_data_single_cell.csv"
os.chdir(folder)
df = pd.read_csv(fname)
```

```python
data = [df["Nloc_488"].to_list(), df["Nloc_647"].to_list()]
labels = ["AF488 probe set 1", "AF647 probe set 2"]
plt.figure(figsize=(6, 4))
ax = plt.hist(
    data,
    label=labels,
    range=(5, 100),
    bins=11,
    histtype="step",
    density=True,
    lw=3,
)
plt.legend(fontsize=18, frameon=False)
plt.xlim(5, 100)
plt.xlabel("#Localization per cluster", fontsize=18)
plt.ylabel("Density", fontsize=18)
plt.gca().spines[:].set_linewidth(1)
plt.gca().tick_params(
    axis="both",
    which="major",
    labelsize=18,
    direction="in",
    bottom=True,
    left=True,
    length=5,
    width=1,
)

plt.savefig("cluster_hist.png", format="png", bbox_inches="tight", dpi=600)
```

```python
plt.figure(figsize=(4, 4))
sns.scatterplot(
    data=df,
    x="Nloc_488",
    y="Nloc_647",
)
plt.xlabel("#Localization per cluster\nAF488 probe set 1", fontsize=18)
plt.ylabel("#Localization per cluster\nAF647 probe set 2", fontsize=18)
plt.gca().spines[:].set_linewidth(1)
plt.axis("equal")
plt.xlim(5, 70)
plt.ylim(5, 70)
plt.gca().tick_params(
    axis="both",
    which="major",
    labelsize=18,
    direction="in",
    bottom=True,
    left=True,
    length=5,
    width=1,
)

plt.savefig("cluster_stoichiometry.png", format="png", bbox_inches="tight", dpi=600)
```
