# HILO Stepwise Photobleach Counting
Run one stage at a time. Open the reported QC PDF before continuing to the next cell.

```python
from pathlib import Path
CONFIG = Path('config.yaml').resolve()
RESUME = True
```

```python
import subprocess
def run_stage(stage):
    command = ['conda','run','-n','smlm','python','run_stepwise_pbsa.py',stage,'--config',str(CONFIG)]
    if RESUME and stage not in {'inspect-inputs','quickpbsa-threshold-pilot','generate-reports'}: command.append('--resume')
    subprocess.run(command, check=True)
```

## 1. Input inspection
Review `01_input_inspection/input_inspection_QC_report.pdf`.

```python
run_stage('inspect-inputs')
```

## 2. Drift correction
Review `02_drift_correction/drift_correction_QC_report.pdf` before detecting ROIs.

```python
run_stage('drift-correction')
```

## 3. ROI detection and growth
Review `03_roi_detection_and_growth/roi_detection_and_growth_QC_report.pdf`.

```python
run_stage('detect-and-grow-rois')
```

## 4. Background-corrected traces
Review `04_background_corrected_traces/background_corrected_trace_QC_report.pdf`.

```python
run_stage('extract-background-corrected-traces')
```

## 5. quickPBSA threshold pilot
Review the pilot PDF, then enter reviewed acquisition-profile thresholds in `config.yaml`.

```python
run_stage('quickpbsa-threshold-pilot')
```

## 6. Photobleaching step counting
This stage refuses to run until all acquisition-profile thresholds are configured.

```python
run_stage('count-photobleaching-steps')
```

## 7. Final reports

```python
run_stage('generate-reports')
```
