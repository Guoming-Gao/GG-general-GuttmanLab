## Run Design

This cell uses `rich.progress` so notebook users can see gene-level progress while the pipeline runs.

```python
import time
import pandas as pd
from rich.console import Console
from rich.progress import Progress, SpinnerColumn, BarColumn, TextColumn, TimeElapsedColumn, TimeRemainingColumn

from utils.designer import design_snpaware_rna_primers


def format_duration(seconds):
    if seconds is None:
        return "--"
    seconds = int(max(0, seconds))
    hours, remainder = divmod(seconds, 3600)
    minutes, seconds = divmod(remainder, 60)
    if hours:
        return f"{hours}h {minutes:02d}m"
    if minutes:
        return f"{minutes}m {seconds:02d}s"
    return f"{seconds}s"


console = Console()
batch_rows = []
run_start = time.monotonic()
completed_gene_durations = []

with Progress(
    SpinnerColumn(),
    TextColumn("[progress.description]{task.description}"),
    BarColumn(),
    TextColumn("{task.completed}/{task.total}"),
    TimeElapsedColumn(),
    TimeRemainingColumn(),
    TextColumn("avg/gene {task.fields[avg_gene]}"),
    console=console,
) as progress:
    gene_task = progress.add_task(
        "Genes",
        total=len(GENE_NAMES),
        avg_gene="--",
    )
    current_task = None

    def make_progress_callback(gene):
        state = {"task": None, "mode": ""}

        def callback(event, **info):
            if event == "mode_start":
                if state["task"] is not None:
                    progress.remove_task(state["task"])
                state["mode"] = info.get("mode", "")
                state["task"] = progress.add_task(
                    f"{gene}: {info.get('message', state['mode'])}",
                    total=max(1, int(info.get("total_attempts", 1))),
                    avg_gene="--",
                )
            elif event == "attempt_start" and state["task"] is not None:
                progress.update(
                    state["task"],
                    description=f"{gene}: {info.get('message', state['mode'])}",
                )
            elif event == "attempt_done" and state["task"] is not None:
                progress.update(
                    state["task"],
                    completed=min(int(info.get("attempt_index", 0)), int(info.get("total_attempts", 1))),
                )
            elif event == "mode_done" and state["task"] is not None:
                progress.update(
                    state["task"],
                    completed=int(info.get("total_attempts", 1)),
                    description=f"{gene}: {state['mode']} complete",
                )
        return callback

    for gene in GENE_NAMES:
        output_dir = OUTPUT_DIR_BY_GENE.get(gene, OUTPUT_ROOT / f"{gene}_SNPaware_RNA_primers")
        done = len(completed_gene_durations)
        avg_gene = sum(completed_gene_durations) / done if done else None
        progress.update(
            gene_task,
            description=f"Gene {done + 1}/{len(GENE_NAMES)}: {gene}",
            avg_gene=format_duration(avg_gene),
        )
        gene_start = time.monotonic()
        try:
            result = design_snpaware_rna_primers(
                gene,
                output_dir,
                dict(CONFIG),
                progress_callback=make_progress_callback(gene),
            )
            gene_elapsed = time.monotonic() - gene_start
            completed_gene_durations.append(gene_elapsed)
            batch_rows.append({
                "Gene": gene,
                "Status": "success",
                "Output_Dir": str(output_dir),
                "Selected_Transcript": result["selected_transcript"].transcript_id,
                "Common_Junctions": len(result["common_junctions"]),
                "Candidate_Pairs": len(result["candidate_rows"]),
                "Top_Pairs": len(result["top_rows"]),
                "Selected_Filter_Tier": result["filter_info"]["selected_tier"],
                "Fallback_Used": result["fallback_used"],
                "Best_Left_Primer": result["top_rows"][0]["Left_Primer_Seq"] if result["top_rows"] else "",
                "Best_Right_Primer": result["top_rows"][0]["Right_Primer_Seq"] if result["top_rows"] else "",
                "Elapsed": format_duration(gene_elapsed),
                "Error": "",
            })
        except Exception as exc:
            gene_elapsed = time.monotonic() - gene_start
            completed_gene_durations.append(gene_elapsed)
            batch_rows.append({
                "Gene": gene,
                "Status": "failed",
                "Output_Dir": str(output_dir),
                "Selected_Transcript": "",
                "Common_Junctions": 0,
                "Candidate_Pairs": 0,
                "Top_Pairs": 0,
                "Selected_Filter_Tier": "failed",
                "Fallback_Used": "",
                "Best_Left_Primer": "",
                "Best_Right_Primer": "",
                "Elapsed": format_duration(gene_elapsed),
                "Error": str(exc),
            })
            console.print(f"[red]{gene} failed:[/red] {exc}")

        avg_gene = sum(completed_gene_durations) / len(completed_gene_durations)
        progress.update(
            gene_task,
            advance=1,
            description=f"Completed {gene}" if progress.tasks[gene_task].completed < len(GENE_NAMES) else "Completed all genes",
            avg_gene=format_duration(avg_gene),
        )

batch_summary = pd.DataFrame(batch_rows)
OUTPUT_ROOT.mkdir(parents=True, exist_ok=True)
batch_summary.to_csv(OUTPUT_ROOT / "SNPaware_RNA_batch_summary.csv", index=False)
console.print(f"[green]Finished {len(batch_rows)} genes in {format_duration(time.monotonic() - run_start)}[/green]")
batch_summary
```
