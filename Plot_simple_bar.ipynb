```python
import os
import numpy as np
import pandas as pd
from scipy.signal import find_peaks
from scipy.stats import sem
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True, style="white")
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
values = [36, 35.2, 0.92, 0, 34.4, 4.72, 0, 25.2, 38.4, 10.8, 0, 27.2]
labels = [
    "control input",
    "control unbound",
    "control wash",
    "control elute",
    "exon input",
    "exon unbound",
    "exon wash",
    "exon elute",
    "intron input",
    "intron unbound",
    "intron wash",
    "intron elute",
]
colors = [
    sns.color_palette()[0],
    sns.color_palette()[0],
    sns.color_palette()[0],
    sns.color_palette()[0],
    sns.color_palette()[1],
    sns.color_palette()[1],
    sns.color_palette()[1],
    sns.color_palette()[1],
    sns.color_palette()[2],
    sns.color_palette()[2],
    sns.color_palette()[2],
    sns.color_palette()[2],
]

plt.figure(figsize=(5, 4))
plt.bar(
    labels,
    values,
    color=colors,
)
plt.ylabel("Concentration, ng/ul", fontsize=15)
plt.gca().spines[:].set_linewidth(1)
plt.title("Bead binding test, Qbit quantification", fontsize=15)
plt.gca().tick_params(
    axis="both",
    which="major",
    labelsize=15,
    direction="in",
    bottom=True,
    left=True,
    length=5,
    width=1,
    rotation=90,
)
plt.savefig("simplebar-beadwashtest-20250522-qbitall.png", format="png", bbox_inches="tight", dpi=300)
```
