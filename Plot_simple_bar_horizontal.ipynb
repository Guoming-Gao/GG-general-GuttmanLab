```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True, style="white")
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
category_names = [
    "mono",
    "bi",
]
results = {
    "wt": [158, 4],
    "d": [213, 15],
    "wt, dox": [60, 1],
    "d, dox": [23, 11],
}

labels = list(results.keys())
data = np.array(list(results.values()))
data = data / np.repeat(data.sum(axis=1), 2).reshape(4, 2) * 100
data_cum = data.cumsum(axis=1)
category_colors = ["#76777b", "#b22222"]

fig, ax = plt.subplots(figsize=(6, 4))
ax.invert_yaxis()
ax.set_xlim(0, np.sum(data, axis=1).max())

for i, (colname, color) in enumerate(zip(category_names, category_colors)):
    widths = data[:, i]
    starts = data_cum[:, i] - widths
    rects = ax.barh(labels, widths, left=starts, height=0.5, label=colname, color=color)
    annotations = ax.bar_label(
        rects,
        color="black",
        fontsize=17,
        fmt="{:.1f}",
        label_type="center",
        padding=1,
    )
    for annotation in annotations:
        x, y = annotation.get_position()
        annotation.set_position((x, y + 23))

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_linewidth(3)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
plt.tight_layout()
plt.savefig("PrelimData_panelA_stackbar.pdf", format="pdf")
```
